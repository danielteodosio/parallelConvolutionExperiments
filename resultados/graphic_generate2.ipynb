```python
import matplotlib.pyplot as plt
import seaborn as sns

# Dados para o primeiro subplot
x1 = [1, 2, 3, 4]
y1 = [927.99, 940.74, 953.53, 965.97]

# Dados para o segundo subplot
x2 = [1, 2, 3, 4]
y2 = [43.333, 75.333, 89.8, 116.416]

# Criar a figura e os subplots
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 5))

# Primeiro subplot
ax1.plot(x1, y1, 'k--*', markersize=10)
ax1.set_title('Mdia de memria alocada - Dataset H4ac')
ax1.set_xlabel('Nmero de segmentos')
ax1.set_ylabel('Memria alocada (MB)')
ax1.set_ylim(927, 968)
ax1.set_xticks([1, 2, 3, 4])

# Segundo subplot
ax2.plot(x2, y2, 'r--*', markersize=10)
ax2.set_title('Mdia de tempo por poca de treinamento - Dataset H4ac')
ax2.set_xlabel('Nmero de segmentos')
ax2.set_ylabel('Tempo mdio por poca (s)')
ax2.set_ylim(41, 119)
ax2.set_xticks([1, 2, 3, 4])

# Ajustar o layout
plt.tight_layout()

# Mostrar o grfico
plt.show()
```
