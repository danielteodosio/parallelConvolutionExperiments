```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Dados para o primeiro subplot
data1 = {
    'categoria': ['Sequencial'] * 4 + ['Paralela'] * 3,
    'Valores': [86.63, 86.73, 87.243, 86.424, 86.9857142857143, 86.967, 87.328],
    'Nomes': ['$S_{1}$', '$S_{3}$', '$S_{6}$', '$S_{9}$', '$P_{1, 3}$', '$P_{1, 3, 6}$', '$P_{1, 3, 6, 9}$']
}

df1 = pd.DataFrame(data1)

# Dados para o segundo subplot
data2 = {
    'categoria': ['Sequencial'] * 4 + ['Paralela'] * 3,
    'Valores': [20, 25, 17, 32, 12, 28, 24],
    'Nomes': ['$S_{1}$', '$S_{3}$', '$S_{6}$', '$S_{9}$', '$P_{1, 3}$', '$P_{1, 3, 6}$', '$P_{1, 3, 6, 9}$']
}

df2 = pd.DataFrame(data2)

# Criação dos subplots
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 7))  # Altura reduzida pela metade

# Primeiro subplot
sns.barplot(ax=ax1, x='Nomes', y='Valores', hue='categoria', data=df1, palette=['blue', 'orange'], dodge=False)
ax1.set_xlabel('Experimentos')
ax1.set_ylabel('% média de acurácia')
ax1.set_title('Resultados de acurácia média - dataset H4')
ax1.legend(title='Tipo de arquitetura', loc='upper left')
ax1.set_yticks(range(86, 88, 1))

# Segundo subplot
sns.barplot(ax=ax2, x='Nomes', y='Valores', hue='categoria', data=df2, palette=['blue', 'orange'], dodge=False)
ax2.set_xlabel('Experimentos')
ax2.set_ylabel('% média de acurácia')
ax2.set_title('Resultados de acurácia média - dataset H4ac')
ax2.legend(title='Tipo de arquitetura', loc='upper left')
ax2.set_yticks(range(0, 101, 5))

# Ajuste do layout
plt.tight_layout()

# Exibição do gráfico
plt.show()

```
