```python
import numpy as np
import matplotlib.pyplot as plt

# Funes de ativao
def step(x):
    return np.where(x<0, 0, 1)

def linear(x):
    return x

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def tanh(x):
    return np.tanh(x)

def relu(x):
    return np.maximum(0, x)

def leaky_relu(x):
    return np.where(x > 0, x, x * 0.07)

# Criao do grid
fig, axs = plt.subplots(3, 2, figsize=(10, 15))

# Dados para plotar
x = np.linspace(-10, 10, 400)
functions = [(step, 'Degrau', 'f(x) = 1, se x >= 0; 0, se x < 0'), 
             (linear, 'Linear', 'f(x) = x'), 
             (sigmoid, 'Sigmoide', 'f(x) = 1 / (1 + e^-x)'), 
             (tanh, 'Tangente Hiperblica', 'f(x) = tanh(x)'), 
             (relu, 'Relu', 'f(x) = max(0, x)'), 
             (leaky_relu, 'Leaky Relu', 'f(x) = x, se x > 0; ax, se x <= 0')]

# Plotagem das funes
for i, ax in enumerate(axs.flatten()):
    func, title, formula = functions[i]
    ax.plot(x, func(x))
    ax.set_title(f'{chr(97 + i)}. {title}')
    ax.text(0.05, 0.99, formula, horizontalalignment='left', verticalalignment='top', transform=ax.transAxes)
    ax.grid()

plt.tight_layout()
plt.show()
```
